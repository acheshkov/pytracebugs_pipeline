## 2. Representator usage

```python
text1 = 'x = a + b'
text2 = 'x = a - b'
text = [text1, text2]
```

```python
e = representator.run(source_code=text)
print(e.shape)
```

```python
t = representator.back(embeddings=torch.tensor(e))
print(t)
```

```python
texts = representator.to_text(t)
print(texts)
```
