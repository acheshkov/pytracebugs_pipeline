```python
from representator import Representator
from vae import VAESimplex, VAEMultiplex
```

## Settings

## 1. VAE definition

```python
vae = VAESimplex(input_size=768)
```

```python
vaem = VAEMultiplex(input_shape=(512, 768))
```

## 2. Representator definition

```python
representator = Representator()
```

## 3. Representator usage

```python
text1 = 'x = a + b'
text2 = 'x = a - b'
text = [text1, text2]
```

```python
e = representator.run(source_code=text, pad=512, pad_random=True)
print(e.shape)
```

```python
t = representator.back(embeddings=e)
print(t)
```

```python
texts = representator.to_text(t)
print(texts)
```

## 4. VAE usage

```python
e.shape
```

```python
e_mean = torch.mean(e, 1)
print(e_mean.shape)
```

```python
n, _, _ = vae(e_mean)
print(n.shape)
```

```python
x = torch.ones(size=(2, 1, 512, 768), requires_grad=False)
print(x.shape)
```

```python
n, _, _ = vaem(x)
print(n.shape)
```

```python
m, _, _ = vaem(e.unsqueeze(1))
print(m.shape)
```
