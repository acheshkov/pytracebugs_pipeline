## Iterate rows

```python
with open('bugs_report.txt', 'w') as f:

  for index, row in df.iterrows():
    bug = row['before_merge']
    fix = row['after_merge']

    f.write('\n\n')
    f.write('*'*30 + '\n')
    f.write(str(index) + '\n')
    f.write('*'*30 + '\n')
    f.write(bug + '\n')
    f.write('.'*30 + '\n' )
    f.write(fix + '\n' )
    f.write('.'*30 + '\n')
    [f.write(d+ '\n') for d in difflib.unified_diff(bug.split('\n'), fix.split('\n'))]
    '''
    print('*'*30)
    print(index)
    print('*'*30)
    print(bug)
    print('-'*30)
    print(fix)
    print('-'*30)
    [print(d) for d in difflib.unified_diff(bug.split('\n'), fix.split('\n'))]
    '''
```
