```python
import pandas as pd
import collections
import pickle
```

# Загрузка данных по коду с ошибками

Загрузим данные всех датасетов

```python
source_code_data_train = pd.read_pickle("filtered_traceback_bugfixes_train.pickle")
source_code_data_valid = pd.read_pickle("filtered_traceback_bugfixes_valid.pickle")
source_code_data_test = pd.read_pickle("bugfixes_test.pickle")
```

Создадим колонку с названием репозитория, содержащего каждый сниппет.

```python
source_code_data_train['repo_name'] = source_code_data_train['url'].apply(lambda x: '/'.join(x.split('/')[3:5]))
source_code_data_valid['repo_name'] = source_code_data_valid['url'].apply(lambda x: '/'.join(x.split('/')[3:5]))
source_code_data_test['repo_name'] = source_code_data_test['url'].apply(lambda x: '/'.join(x.split('/')[3:5]))
```

```python
def load_repo_names(dataset):
    return list(set(dataset["repo_name"].to_list()))
```

```python
train_repos = load_repo_names(source_code_data_train)
valid_repos = load_repo_names(source_code_data_valid)
test_repos = load_repo_names(source_code_data_test)
```

Посмотрим сколько репозиториев представлено в коде с ошибками.

```python
len(test_repos), len(train_repos), len(valid_repos), len(test_repos) + len(train_repos) + len(valid_repos)
```

Посмотрим сколько сниппетов лежит в каждом репозитории

```python
train_repos_count = source_code_data_train['repo_name'].value_counts()

for i, count in zip(train_repos_count.index, train_repos_count):
    print(i, count)
```

```python
valid_repos_count = source_code_data_valid['repo_name'].value_counts()

for i, count in zip(valid_repos_count.index, valid_repos_count):
    print(i, count)
```

```python
test_repos_count = source_code_data_test['repo_name'].value_counts()

for i, count in zip(test_repos_count.index, test_repos_count):
    print(i, count)
```

Сниппеты обучающей и тестовой выборок не пересекаются по репозиториям:

```python
set(test_repos) & set(train_repos)
```

Репозитории в валидационной и тестовой выборках совпадают:

```python
set(test_repos) == set(valid_repos)
```

# Топики репозиториев

Загрузим данные с топиками репозиториев для кода с ошибками.

```python
repo_name_to_topics = pickle.load(open("topics_slashes.pickle", "rb"))
train_repos_topics = [repo_name_to_topics[repo_name] for repo_name in source_code_data_train['repo_name'].to_numpy() if repo_name in repo_name_to_topics]
valid_repos_topics = [repo_name_to_topics[repo_name] for repo_name in source_code_data_valid['repo_name'].to_numpy()  if repo_name in repo_name_to_topics]
test_repos_topics = [repo_name_to_topics[repo_name] for repo_name in source_code_data_test['repo_name'].to_numpy()  if repo_name in repo_name_to_topics]
```

```python
train_repos_topics_cnt = collections.Counter()
for t in train_repos_topics:
    train_repos_topics_cnt.update(set(t))
    
valid_repos_topics_cnt = collections.Counter()
for t in valid_repos_topics:
    valid_repos_topics_cnt.update(set(t))

test_repos_topics_cnt = collections.Counter()
for t in test_repos_topics:
    test_repos_topics_cnt.update(set(t))
```

Наиболее представленные топики в обучающей выборке:

```python
train_repos_topics_cnt.most_common()
```

Наиболее представленные топики в валидационной выборке:

```python
valid_repos_topics_cnt.most_common()
```

Наиболее представленные топики в тестовой выборке:

```python
test_repos_topics_cnt.most_common()
```

Сравним представленности топиков:

```python
train_topics_count = pd.DataFrame(train_repos_topics_cnt.most_common())
train_topics_count.rename({0: 'topic name', 1: 'train count'},
                           inplace=True,
                           axis=1)
train_topics_count.set_index(keys='topic name', inplace=True)

valid_topics_count = pd.DataFrame(valid_repos_topics_cnt.most_common())
valid_topics_count.rename({0: 'topic name', 1: 'valid count'},
                           inplace=True,
                           axis=1)
valid_topics_count.set_index(keys='topic name', 
                             inplace=True)
test_topics_count = pd.DataFrame(test_repos_topics_cnt.most_common())
test_topics_count.rename({0: 'topic name', 1: 'test count'},
                           inplace=True,
                           axis=1)
test_topics_count.set_index(keys='topic name', inplace=True)
```

```python
topics_count = pd.concat((train_topics_count, valid_topics_count, test_topics_count), 
                         axis=1)
```

```python
topics_count['train topic ratio'] = topics_count['train count'] / source_code_data_train.shape[0]
topics_count['valid topic ratio'] = topics_count['valid count'] / source_code_data_valid.shape[0]
topics_count['test topic ratio'] = topics_count['test count'] / source_code_data_test.shape[0]
```

```python
topics_count = topics_count.fillna(0)
topics_count
```

Отсортируем по столбцу `train topic ratio`.

```python
topics_count.sort_values(by='train topic ratio',
                         ascending=False).head(30)
```

Во всех выборках хорошо представлен топик python, в train лучше представлен топик machine-learning.

# Сравнение датасетов с ошибками и датасетов с корректным кодом

## На уровне репозиториев

Сравним сколько у обучающих подвыборок из каждого класса имеется общих репозиториев

```python
modified_train_repos = [repo_name.replace('/', '_') for repo_name in train_repos]
modified_valid_repos = [repo_name.replace('/', '_') for repo_name in valid_repos]
modified_test_repos = [repo_name.replace('/', '_') for repo_name in test_repos]
```

```python
len(set(modified_train_repos) & set(train_repos_)), min(len(modified_train_repos), len(train_repos_))
```

Таким образом, имеется 94 репозитория, сниппеты из которых представлены как обучающей выборке кода с ошибками, так и в обучающей выборке с корректным кодом. В обучающей выборке с кодом с ошибками представлено 562 репозитория. Таким образом, 16% репозиториев, представлены в обеих выборках.

Теперь для валидационной выборки:

```python
len(set(modified_valid_repos) & set(valid_repos_)), min(len(modified_valid_repos), len(valid_repos_))
```

Таким образом, имеется 67 репозитория, сниппеты из которых представлены как валидационной выборке кода с ошибками, так и в валидационной выборке с корректным кодом. В валидационной выборке с кодом с ошибками представлено 68 репозиториев. Таким образом, почти 100% репозиториев, представлены в обеих выборках.

Теперь для тестовой выборки:

```python
len(set(modified_test_repos) & set(test_repos_)), min(len(modified_test_repos), len(test_repos_))
```

Таким образом, имеется 1 репозиторий, сниппеты из которых представлены как тестовой выборке кода с ошибками, так и в тестовой выборке с корректным кодом. В тестовой выборке с кодом с ошибками представлено 68 репозиториев. Таким образом, только 1.5% репозиториев представлены в обеих выборках.
