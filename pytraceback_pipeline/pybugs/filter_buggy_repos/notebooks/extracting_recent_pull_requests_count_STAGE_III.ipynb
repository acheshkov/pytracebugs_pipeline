```python
from collect_repos import create_repos_dataset_with_recent_PR_count
```

```python
repos_info = pd.read_csv('../data/interesting_repos_info.csv')
repos_info = repos_info.astype({'forkCount': 'int32', 'stargazerCount': 'int32',
                                'diskUsage': 'float64', 'commitComments.totalCount': 'int32',
                                'watchers.totalCount': 'int32', 'issues.totalCount': 'int32',
                                'pullRequests.totalCount': 'int32', 'releases.totalCount': 'int32',
                                'databaseId': 'int32', 'mirrorUrl':'str', 'parent': 'str'})
```

```python
repos_info.describe()
```

```python
quantile_threshold = 0.25
repos_info['max_of_refs_count_n_releases_count'] = repos_info[['refs.totalCount', 
                                                               'releases.totalCount']].max(axis=1)
print(f'Tags count {quantile_threshold}th percentile value={repos_info["max_of_refs_count_n_releases_count"].quantile(q=quantile_threshold)}')
```

```python
repos_info_with_no_tags = repos_info.loc[repos_info['max_of_refs_count_n_releases_count']==0]
repos_info_with_no_tags_n_many_issues = repos_info_with_no_tags.loc[repos_info_with_no_tags['issues.totalCount']>=repos_info_with_no_tags['issues.totalCount'].quantile(q=0.99)]
repos_info_with_no_tags_n_many_issues[['nameWithOwner', 'issues.totalCount', 'stargazerCount']]
```

```python
repos_info_with_1_tag = repos_info.loc[repos_info['max_of_refs_count_n_releases_count']==1]
repos_with_1_tag_n_many_issues = repos_info_with_1_tag.loc[repos_info_with_1_tag['issues.totalCount'] >= repos_info_with_1_tag['issues.totalCount'].quantile(q=0.99)]
repos_with_1_tag_n_many_issues[['nameWithOwner', 'issues.totalCount', 'stargazerCount']]
```

```python
filtered_repos_info = repos_info.loc[repos_info['refs.totalCount']>=repos_info["refs.totalCount"].quantile(q=quantile_threshold)]
```
