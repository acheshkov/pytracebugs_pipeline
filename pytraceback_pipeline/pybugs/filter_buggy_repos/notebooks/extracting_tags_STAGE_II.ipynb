```python
from collect_repos import create_repos_dataset_with_tags, load_repo_info_from_json
```

```python
path_to_repos_list = '../data/'
repos_list_filename = 'python_repos_interesting.json'
```

```python
repos_info = load_repo_info_from_json(path_to_repos_list,
                                      repos_list_filename)
repos_info = repos_info.astype({'forkCount': 'int32', 'stargazerCount': 'int32',
                                'diskUsage': 'float64', 'commitComments.totalCount': 'int32',
                                'watchers.totalCount': 'int32', 'issues.totalCount': 'int32',
                                'pullRequests.totalCount': 'int32', 'releases.totalCount': 'int32',
                                'databaseId': 'int32', 'mirrorUrl':'str', 'parent': 'str',
                                'licenseInfo':'str'})
```

```python
repos_info.describe()
```

```python
quantile_threshold = 0.75
print(f'Stargazers count {quantile_threshold}th percentile value={repos_info["stargazerCount"].quantile(q=quantile_threshold)}')
print(f'Issues count {quantile_threshold}th percentile value={repos_info["issues.totalCount"].quantile(q=quantile_threshold)}')
print(f'PullRequests count {quantile_threshold}th percentile value={repos_info["pullRequests.totalCount"].quantile(q=quantile_threshold)}')
```

```python
filtered_repos_info = repos_info.loc[
    (repos_info['stargazerCount'] >= repos_info['stargazerCount'].quantile(q=quantile_threshold)) &
    (repos_info['issues.totalCount'] >= repos_info['issues.totalCount'].quantile(q=quantile_threshold)) &
    (repos_info['pullRequests.totalCount'] >= repos_info['pullRequests.totalCount'].quantile(q=quantile_threshold))
]
```
