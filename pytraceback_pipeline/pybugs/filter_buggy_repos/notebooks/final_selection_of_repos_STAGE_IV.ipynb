```python
import pandas as pd
```

```python
repos_info = pd.read_csv('../data/further_extended_interesting_repos_info.csv')
repos_info = repos_info.astype({'forkCount': 'int32', 'stargazerCount': 'int32',
                                'diskUsage': 'float64', 'commitComments.totalCount': 'int32',
                                'watchers.totalCount': 'int32', 'issues.totalCount': 'int32',
                                'pullRequests.totalCount': 'int32', 'releases.totalCount': 'int32',
                                'databaseId': 'int32', 'mirrorUrl':'str', 'parent': 'str'})
```

```python
issues_count_quantile_threshold = 0.8
stargazer_count_quantile_threshold = 0.5
print(f'Recent (within 180 days) pull requests count {issues_count_quantile_threshold}th percentile value={repos_info["issueCount"].quantile(q=issues_count_quantile_threshold)}')
print(f'Stargazer count {stargazer_count_quantile_threshold}th percentile value={repos_info["stargazerCount"].quantile(q=stargazer_count_quantile_threshold)}')
```

```python
filtered_repos_info = repos_info.loc[(repos_info['issueCount'] >= repos_info['issueCount'].quantile(q=issues_count_quantile_threshold)) &
                                     (repos_info['stargazerCount'] >= repos_info['stargazerCount'].quantile(q=stargazer_count_quantile_threshold))]
```

```python
 filtered_repos_info.describe()
```

```python
filtered_repos_info.shape
```

```python
filtered_repos_info.to_csv('../data/big_list_of_repos.csv', index=False)
```
