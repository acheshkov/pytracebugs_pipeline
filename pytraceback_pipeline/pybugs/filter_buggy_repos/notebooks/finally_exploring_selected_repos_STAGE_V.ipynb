```python
import pandas as pd
from form_issues_datasets import infer_bug_label
```

```python
filtered_repos = pd.read_csv('../data/big_list_of_repos.csv')
filtered_repos.set_index('nameWithOwner', inplace=True)
filtered_repos['bug_labels'] = [{'labels': []} for i in range(filtered_repos.shape[0])]
```

```python
for repo_index, repo_name in enumerate(filtered_repos.index):
    print(repo_index, repo_name)
    filtered_repos.at[repo_name, 'bug_labels']['labels'] = infer_bug_label(repo_name)
```

```python
filtered_repos['bug_labels'] = filtered_repos['bug_labels'].apply(lambda x: x['labels'])
filtered_repos.to_csv('../data/big_list_of_repos_with_bug_labels_1.csv')
```

```python
all_bug_labels = []
for bug_labels in filtered_repos['bug_labels']:
    all_bug_labels += bug_labels
```

```python
from collections import Counter
all_bug_labels_counted = Counter(all_bug_labels)
for label, count in all_bug_labels_counted.items():
    print(f'Inferred bug label: "{label}". It occurs in {count} repos')
```
